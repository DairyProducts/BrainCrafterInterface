```python
# Load numpy arrays
SESSION_NUMBER = 7
aux_continuous = np.load(f"{DATA_FOLDER}{DATA_SESSION}aux-continuous-session-{SESSION_NUMBER}.npy", allow_pickle=True)
baseline = np.load(f"{DATA_FOLDER}{DATA_SESSION}baseline-session-{SESSION_NUMBER}.npy", allow_pickle=True)
eeg_continuous = np.load(f"{DATA_FOLDER}{DATA_SESSION}eeg-continuous-session-{SESSION_NUMBER}.npy", allow_pickle=True)
filtered_baseline = np.load(f"{DATA_FOLDER}{DATA_SESSION}filtered-baseline-session-{SESSION_NUMBER}.npy", allow_pickle=True)
filtered_session = np.load(f"{DATA_FOLDER}{DATA_SESSION}filtered-session-{SESSION_NUMBER}.npy", allow_pickle=True)
raw_session = np.load(f"{DATA_FOLDER}{DATA_SESSION}raw-session-{SESSION_NUMBER}.npy", allow_pickle=True)
```

```python
# Pair channels numbers with names
channels = {
    0: "F3",
    1: "F4",
    2: "C3",
    3: "Cz",
    4: "C4",
    5: "P3",
    6: "Pz",
    7: "P4",
    "GND": "GND",
    "REF": "Fz"
}
```

```python
print(filtered_baseline.shape)
print(filtered_session.shape)
print(filtered_session[0, 1].shape)
```

```python
TRIAL = 7

fig, axes = plt.subplots(8, 1, figsize=(12, 16), sharex=True)

for i in range(filtered_session[TRIAL, 1].shape[0]):
    axes[i].plot(filtered_session[TRIAL, 1][i])
    axes[i].set_ylabel(f"Channel {channels[i]}, index {i}")

fig.suptitle(f"Trial {TRIAL}, Action: {filtered_session[TRIAL, 0]}")
axes[-1].set_xlabel("Sample")
plt.tight_layout()
plt.show()
```
